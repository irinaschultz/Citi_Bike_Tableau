```python
#Dependencies
import pandas as pd
```

```python
# List of file names to be combined  for 2018
data_18 = ['data/201801.csv', 'data/201802.csv', 'data/201803.csv', 'data/201804.csv', 'data/201805.csv', 'data/201806.csv', 'data/201807.csv', 'data/201808.csv', 'data/201809.csv', 'data/201810.csv', 'data/201811.csv', 'data/201812.csv']

# Create an empty list to store each dataframe for 2018
df_2018 = []
```

```python
# Loop through each file in the list for 2018
for data in data_18:
    # Read the file into a dataframe
    df18 = pd.read_csv(data)

    # Append the dataframe to the list for 2018
    df_2018.append(df18)

```

```python
# Show dataframe 2018
df18
```

```python
# List of file names to be combined  for 2019
data_19 = ['data/201901.csv', 'data/201902.csv', 'data/201903.csv', 'data/201904.csv', 'data/201905.csv', 'data/201906.csv', 'data/201907.csv', 'data/201908.csv', 'data/201909.csv', 'data/201910.csv', 'data/201911.csv', 'data/201912.csv']

# Create an empty list to store each dataframe for 2019
df_2019 = []
```

```python
# Loop through each file in the list for 2019
for data in data_19:
    # Read the file into a dataframe
    df19 = pd.read_csv(data)

    # Append the dataframe to the list for 2019
    df_2019.append(df19)
```

```python
# Show dataframe 2019
df19
```

```python
# List of file names to be combined  for 2020
data_20 = ['data/202001.csv', 'data/202002.csv', 'data/202003.csv', 'data/202004.csv', 'data/202005.csv', 'data/202006.csv', 'data/202007.csv', 'data/202008.csv', 'data/202009.csv', 'data/202010.csv', 'data/202011.csv', 'data/202012.csv']

# Create an empty list to store each dataframe for 2020
df_2020 = []
```

```python
# Loop through each file in the list for 2020
for data in data_20:
    # Read the file into a dataframe
    df20 = pd.read_csv(data)

    # Append the dataframe to the list for 2019
    df_2020.append(df20)
```

```python
# Show dataframe 2019
df20
```

```python
# Concatenate all dataframes for all the years into one
three_years_df = pd.concat([df18, df19, df20], axis=0)

# Display the results
three_years_df
```

```python
# Dictionary mapping values in the 'gender' column to new values
mapping = {0: 'Unknown', 1: 'Male', 2: 'Female'}

# Use the 'map' method to change the values in the 'gender' column
three_years_df['gender'] = three_years_df['gender'].map(mapping)

# Show DataFrame
three_years_df
```

```python
# Save the dataframe to a CSV file
three_years_df.to_csv('data/citibike3.csv', index=False)
```
